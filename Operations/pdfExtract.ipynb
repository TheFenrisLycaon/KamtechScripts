```python
import tabula
import pandas as pd
import textract
import datefinder
```

```python
df = tabula.read_pdf('typeII.pdf', pages='all', guess=True, lattice=True)
```

```python
fileNO = 1
for i in range(len(df)):
    k = pd.DataFrame(df[i])
    k.replace(r'\s', ' ', regex=True, inplace=True)
    if len(k) != 0:
        with open(f'data/{fileNO}.json', 'w', encoding='utf-8') as file:
                k.to_json(file, force_ascii=False)
                k.to_csv(f'data/{fileNO}.csv', encoding='utf-8', index=False)
                fileNO+=1
```
